```python
#функция обогащения
def enrichment(rate_input,rate_output,rate_tail):
    return(100 * rate_output*(rate_input - rate_tail)/ rate_input / (rate_output - rate_tail))

df_full['check_recovery']= enrichment(df_full['rougher.input.feed_au'], 
                                      df_full['rougher.output.concentrate_au'],
                                      df_full['rougher.output.tail_au'] )
print('Среднее линейное отклонение расчетного и исходного коэффициента обогащения:',
      mean_absolute_error(df_full['rougher.output.recovery'],df_full['check_recovery']) )
```

## Модель

### Подготовка к расчету
Для расчета исходные датасеты не подходят. В тестовом датасете отсутствуют ряд параметров, которые имеются в тренировочном датафрейме, вследствие чего модель будет обучена с данными, которые потом будут отсутствовать. Это приведет к ошибкам, поэтому необходимо иметь тренировочный датасет близкий к тестовому. Так же требуется предсказать две величины - 'rougher.output.recovery' и 'final.output.recovery'. Для этого потребуется обучить разные модели для каждого признака. 

Так же на этапе исследования было выявлено, что признаки имеют разный масштаб, поэтому стандартизуем их с помощью StandartScaler, что даст улучшение показателей моделей.

```python
target_rgh = df_train['rougher.output.recovery']
target_fin = df_train['final.output.recovery']

features = df_train[df_test.columns] #извлекаем в features аналогичные тестовому датафрейму колонки

scaler = StandardScaler() #стандартизатор 
features = scaler.fit_transform(features) #стандартизация
```

Обучим модели, которые подходят для задачи регрессии: случайный лес и логистическую регрессию. Дерево решений не будем отдельно рассматривать, т.к. оно близко по сути к случайному лесу с маленьким количеством деревьев. Оценим результаты работы кросс-валидацией. 
### Модель случайного леса

Рассмотрим модель случайного леса. Подбирать гиперпараметры особо не требуется, т.к. решение быстро сходится к одной величине. 

```python
def smape(target,prediction):
    mn = abs(target-prediction)*200/( abs(target)+abs(prediction) )
    return( mn.mean() )

score_smape = make_scorer(smape, greater_is_better=True) # создаем свой оценщик

# инициализация лучшей метрики
best_score_rgh=100 

# лучшие параметры случайного леса
n_est_best = 0
n_depth_best = 0
print('Средние оценки на этапе флотации при кросс-валидации:') 
for depth in range(5, 26, 5): 
    for estim in range(5, 16, 5):
        model = RandomForestRegressor(n_estimators=estim, max_depth = depth,n_jobs=-1, random_state=12345)
        model.fit(features, target_rgh)
        scores = cross_val_score(model, features, target_rgh,scoring=score_smape, cv=5).mean()
        print("Кол-во деревьев:", estim,"  глубина:",depth ,"  ср.sMAPE", scores)
        
        #ищем лучшую модель
        if scores<best_score_rgh:
            best_score_rgh = scores
            n_est_best = estim
            n_depth_best = depth
            mod_rf_rgh = model

print('-лучший результат-')
print("Кол-во деревьев:",n_est_best,"  глубина:", n_depth_best,"  ср.sMAPE", best_score_rgh)
```

```python
# лучшая метрика
best_score_fin=100 

# лучшие параметры случайного леса
n_est_best = 0
n_depth_best = 0
print('Средние оценки на этапе очистки при кросс-валидации:') 
for depth in range(5, 26, 5): 
    for estim in range(5, 16, 5):
        model = RandomForestRegressor(n_estimators=estim, max_depth = depth, random_state=12345)
        model.fit(features, target_fin)
        scores = cross_val_score(model, features, target_rgh,scoring=score_smape, n_jobs=-1, cv=5).mean()
        print("Кол-во деревьев:", estim,"  глубина:",depth ,"  ср.sMAPE", scores)
        
        #ищем лучшую модель 
        if scores<best_score_fin:
            best_score_fin = scores
            n_est_best = estim
            n_depth_best = depth
            mod_rf_fin = model

print('-лучший результат-')
print("Кол-во деревьев:",n_est_best,"  глубина:", n_depth_best,"  ср.sMAPE", best_score_fin)

print('Для модели случайного леса итоговый sMAPE', 0.25*best_score_rgh + 0.75*best_score_fin )
```

Вывод: погрешность модели на этапе флотации составляет порядка 5%, финальный этап предсказывается с точностью 5%. Итоговый результат комплексной метрики - 5%. 

### Модель линейной регрессии
Рассмотрим модель линейной регрессии с целью получить более высокие результаты метрики sMAPE.

```python
mod_lr_rgh = LinearRegression() #модель для этапа флотации
mod_lr_rgh.fit(features, target_fin)

score_rgh = cross_val_score(mod_lr_rgh, features, target_rgh,scoring=score_smape,n_jobs=-1, cv=5).mean()
print('Средняя оценка sMAPE Для модели линейной регрессии на этапе флотации при кросс-валидации:',score_rgh) 
```

```python
mod_lr_fin = LinearRegression(); #модель для финальной обработки
mod_lr_fin.fit(features, target_fin)

score_fin = cross_val_score(mod_lr_fin, features, target_fin,scoring=score_smape,n_jobs=-1, cv=5).mean()
print('Средняя оценка sMAPE для модели линейной регрессии на этапе очистки при кросс-валидации:',score_fin) 

print('Итоговый sMAPE для линейной регрессии', 0.25*score_rgh + 0.75*score_fin )
```

Модель линейной регрессии показала результаты значительно хуже, вследствие чего для тестового датасета будет применена модель случайного леса.

### Тестирование лучшей модели на тестовой выборке. 

Наилучшие результаты показала модель случайного дерева. Применим эту модель к тестовому датафрейму. 

```python
df_full.set_index('date',drop=True,inplace=True)
targ_test_rgh = df_full.loc[df_test.index,'rougher.output.recovery']
targ_test_fin = df_full.loc[df_test.index,'final.output.recovery']

feat_test = scaler.transform(df_test) #стандартизация 

pred_rgh_test =  mod_rf_rgh.predict(feat_test) #предсказания флотации
score_rgh_test = smape(targ_test_rgh,pred_rgh_test) #метрика на этапе флотации
print("Для модели случайного леса на этапе флотации показатель sMAPE равен", score_rgh_test)
pred_fin_test =  mod_rf_fin.predict(feat_test)  #предсказания окончательной очистики
score_fin_test = smape(targ_test_fin,pred_fin_test)
print("Для модели случайного леса на финальном этапе показатель sMAPE равен", score_fin_test)
print('Итоговый sMAPE', 0.25*score_rgh_test + 0.75*score_fin_test )
```
